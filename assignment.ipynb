- Basic Data Cleaning

```python
df.isnull()
```

```python
df = df[df != "invalid"]
df =  df.dropna()
df
```

```python
df = df.assign(Address=[""]*43)
df
```

```python
del df['Address']
df
```

```python
df.columns = ['Name', 'Age', 'Email Address', 'Country', 'Salary', 'Date Joined']
df
```

```python
def salary_increase(salary):
    increased = str((float(salary)*2))
    return increased

df['Salary'] = df['Salary'].apply(salary_increase)
df
```
